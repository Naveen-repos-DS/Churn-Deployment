# Model Implementation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x = new_df.drop(['Churn_Yes'],axis=1)
y = new_df['Churn_Yes']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=7)
```

```python
df['Churn'].value_counts()
```

```python
from imblearn.over_sampling import SMOTE

smt = SMOTE()
X_train,Y_train = smt.fit_resample(x_train,y_train)
```

```python
Y_train.value_counts()
```

```python

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train,Y_train )
yhat = lr.predict(x_test)

```

```python
yhat
```

```python
from sklearn.metrics import classification_report,confusion_matrix


print(classification_report(y_test,yhat))

```

```python
sns.heatmap(confusion_matrix(y_test, yhat))
```

```python
import numpy as np

from sklearn.ensemble import RandomForestClassifier
```

```python
num_trees = 100
max_feat = int(np.sqrt(X_train.shape[1]))

rfc = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,              # limit depth
    min_samples_split=5,       # avoid too small splits
    min_samples_leaf=2,        # avoid single-node leaves
    max_features='sqrt',
    random_state=7,
    class_weight='balanced'
)
rfc.fit(x_train,y_train)
```

```python
params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0]
}
```

```python
yhat_r = rfc.predict(x_test);yhat_r
```

```python
print(classification_report(y_test,yhat_r))
```

```python
train_accuracy = rfc.score(X_train, Y_train)
print("Training Accuracy:", train_accuracy)
```

```python
test_acc = rfc.score(x_test, y_test)
print("Test Accuracy:", test_acc)
```

```python
feat_imp = pd.Series(rfc.feature_importances_, index=x_train.columns)
print(feat_imp.sort_values(ascending=False).head(10))
```
